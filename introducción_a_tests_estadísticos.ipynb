## Intervalo de confianza

## analizando test estadisticos z test t test

obteniendo el promedio total de las notas


```python
nota_promedio_por_pelicula_10_o_mas_votos.mean()
```

```python
from statsmodels.stats.weightstats import zconfint

zconfint(nota_promedio_por_pelicula_10_o_mas_votos)
```

Se recomienda utilizar el z test si se tiene un volumenes grandes de datos y distribucion normal

existe un 95% de confianza que si selecciono un dato random del universo este este entre (3.4112459477469557, 3.452854733323563)

```python
from statsmodels.stats.weightstats import DescrStatsW


desc_notas_10_votos = DescrStatsW(nota_promedio_por_pelicula_10_o_mas_votos)
desc_notas_10_votos.tconfint_mean()
```

Se obtiene el intervalo de confianza para el test T (3.411234839229381, 3.4528658418411395)

Test T para un numero mas pequeño de muestras

```python
from statsmodels.stats.weightstats import ztest
ztest(notas1.rating,nota_promedio_por_pelicula_10_o_mas_votos)
```

```python
from statsmodels.stats.weightstats import DescrStatsW

descr_todas_las_notas = DescrStatsW(notas.rating)
descr_toy_story = DescrStatsW(notas1.rating)

comparacion = descr_todas_las_notas.get_compare(descr_toy_story)

comparacion.summary()
```

La diferencia de medias entre las calificaciones de todas las notas y las calificaciones de "Toy Story" es -0.4194, con un error estándar de 0.071. El valor del estadístico t es -5.894, y el valor p es 0.000, lo que indica una diferencia altamente significativa entre las dos medias. Además, el intervalo de confianza del 95% para la diferencia de medias (-0.559 a -0.280) no incluye el 0, reforzando la evidencia de que hay una diferencia significativa entre las medias.

al rechazar nuestra hipotesis (nula) consideramos que la media de TS es mayor que la media de las peliculas en general
